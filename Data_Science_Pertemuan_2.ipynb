```python
# ================================================
# PERTEMUAN 2: PERSIAPAN & MEMUAT DATASET
# ================================================
# Nama : Winar Wahyu Wulansari
# NIM : 250401020025
# Kelas : IF405
# ================================================

# 1. Impor library yang dibutuhkan (Sesuai langkah praktikum nomor 4)
import pandas as pd
import numpy as np

# 2. Muat dataset Titanic dari URL GitHub resmi yang diberikan (Sesuai langkah nomor 5)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("--- DATASET BERHASIL DIMUAT ---")
print(f"5 Baris pertama data:\n{df.head()}\n")
```

```python


# ==========================================
# Q1: JUMLAH BARIS, KOLOM, DAN NAMA KOLOM
# ==========================================
print("=== JAWABAN Q1 ===")
# Mengambil dimensi dataframe (baris, kolom)
baris, kolom = df.shape
print(f"Jumlah Baris: {baris}")
print(f"Jumlah Kolom: {kolom}")

# Menampilkan daftar nama kolom yang tersedia
print("Kolom yang tersedia:")
print(list(df.columns))
print("\n")

```

```python
# ==========================================
# Q2: MISSING VALUES (JUMLAH DAN PERSENTASE)
# ==========================================
print("=== JAWABAN Q2 ===")
# Menghitung jumlah missing values per kolom
missing_count = df.isnull().sum()

# Menghitung proporsi persentase missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Menggabungkannya ke dalam satu tabel data agar rapi
missing_data = pd.DataFrame({
    'Jumlah Missing': missing_count,
    'Persentase (%)': missing_percentage
})

# Filter hanya kolom yang memiliki missing values > 0
missing_only = missing_data[missing_data['Jumlah Missing'] > 0]
print(missing_only.round(2))
print("\n")
```

```python
# ==========================================
# Q3: PERSENTASE PENUMPANG YANG SELAMAT
# ==========================================
print("=== JAWABAN Q3 ===")
# Menghitung rata-rata dari kolom 'Survived' (karena bernilai 1 untuk selamat dan 0 untuk tidak)
# dikalikan 100 untuk mendapatkan persentase.
persen_selamat = df['Survived'].mean() * 100
print(f"Persentase penumpang yang selamat: {persen_selamat:.2f}%")
print("\n")
```

```python
# ==========================================
# Q4: FILTER WANITA KELAS 1 (JUMLAH & RATA-RATA USIA)
# ==========================================
print("=== JAWABAN Q4 ===")
# Melakukan filtering dengan kondisi ganda menggunakan operator &
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

jumlah_wanita_kelas1 = len(wanita_kelas1)
rata_usia_wanita_kelas1 = wanita_kelas1['Age'].mean()

print(f"Jumlah penumpang wanita dari Kelas 1: {jumlah_wanita_kelas1} orang")
print(f"Rata-rata usia mereka: {rata_usia_wanita_kelas1:.2f} tahun")
print("\n")
```

```python
# ==========================================
# Q5: TINGKAT KESELAMATAN PER KELAS PENUMPANG
# ==========================================
print("=== JAWABAN Q5 ===")
# Mengelompokkan berdasarkan Pclass dan mencari rata-rata kolom Survived
tingkat_keselamatan = df.groupby('Pclass')['Survived'].mean() * 100
print("Tingkat Keselamatan per Kelas Penumpang (%):")
print(tingkat_keselamatan.round(2))

# Menentukan kelas dengan tingkat keselamatan tertinggi
kelas_tertinggi = tingkat_keselamatan.idxmax()
print(f"Kelas penumpang dengan tingkat keselamatan paling tinggi adalah: Kelas {kelas_tertinggi}")
print("\n")
```

```python
# ==========================================
# Q6: ANALISIS TAMBAHAN (KORELASI TARIF & KESELAMATAN)
# ==========================================
print("=== JAWABAN Q6 ===")
# Analisis tambahan: Melihat rata-rata tarif tiket (Fare) antara penumpang yang selamat vs tidak selamat
tarif_analisis = df.groupby('Survived')['Fare'].mean()
print("Rata-rata Tarif Tiket berdasarkan Status Keselamatan (0 = Wafat, 1 = Selamat):")
print(tarif_analisis.round(2))
```
